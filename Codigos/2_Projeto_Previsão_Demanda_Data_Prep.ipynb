## Funções

```python
# Função para gerar os metadados do dataframe
def generate_metadata(dataframe):
    """
    Gera um dataframe contendo metadados das colunas do dataframe fornecido.

    :param dataframe: DataFrame para o qual os metadados serão gerados.
    :return: DataFrame contendo metadados.
    """

    # Coleta de metadados básicos
    metadata = pd.DataFrame({
        'nome_variavel': dataframe.columns,
        'tipo': dataframe.dtypes,
        'qt_nulos': dataframe.isnull().sum(),
        'percent_nulos': round((dataframe.isnull().sum() / len(dataframe))* 100,2),
        'cardinalidade': dataframe.nunique(),
    })
    metadata=metadata.sort_values(by='percent_nulos',ascending=False)
    metadata = metadata.reset_index(drop=True)

    return metadata
```

## Feature Engineering

### Adicionando colunas de "lag"

```python
# Adicionando colunas de lag para trazer mais variáveis para nossa base
def add_sales_lags(df, group_cols, lags):
  """
  Adiciona colunas de lag no dataframe para períodos específicos.

  Parâmetros:
  :param df: DataFrame de entrada.
  :param group_cols: Lista de colunas para agrupar os dados.
  :param lags: Lista de períodos de lag.

  Returns:
  pd.DataFrame: DataFrame com colunas de lag adicionadas.
  """
  for lag in lags:
    lag_col_name = f'sales_lag_{lag}'
    df[lag_col_name] = df.groupby(group_cols)['sales'].shift(lag)
  return df


# Criando as colunas de lag
group_cols = ['store_nbr', 'family']
lags = [1,3,5,7,14,21,28,30,60,90,120,150,180,360]

df_train_full = add_sales_lags(df_train_full, group_cols, lags)
df_test_full = add_sales_lags(df_test_full, group_cols, lags)
```

### Adicionando colunas de média móvel

```python
def add_sales_rolling_means(df, group_cols, windows):
  """
  Adiciona colunas de média móvel no dataframe para diferentes janelas.

  Parâmetros:
  df: DataFrame de entrada.
  group_cols: Lista de colunas para agrupar os dados.
  windows: Lista de janelas de média móvel.

  Returns:
  pd.DataFrame: DataFrame com colunas de média móvel adicionadas.
  """
  for window in windows:
      rolling_col_name = f'rolling_mean_{window}'
      df[rolling_col_name] = df.groupby(group_cols)['sales'].transform(lambda x: x.rolling(window=window).mean())
  return df


# criando as colunas com as médias móveis
group_cols = ['store_nbr', 'family']
windows = [3,5,7,14,21,28,30,45,60,75,90,120,150,180,360]

df_train_full = add_sales_rolling_means(df_train_full, group_cols, windows)
df_test_full = add_sales_rolling_means(df_test_full, group_cols, windows)
```

### Adicionando variáveis relacionadas a data

```python
# Adicionando features de datas
def add_date_features(df,date_col):
	"""
	Adiciona colunas de ano, mês, dia, dia da semana e se era feriado ou não baseado em uma data indicada.

	Parâmetros:
	df (pd.DataFrame): O dataframe contendo a coluna de data
	date_col (str): o nome da coluna contendo as datas

	Returns:
	pd.DataFrame: Um data frame com as novas colunas adicionadas
	"""
	df['date'] = pd.to_datetime(df['date'])
	df['year'] = df[date_col].dt.year
	df['month'] = df[date_col].dt.month
	df['day'] = df[date_col].dt.day
	df['day_of_week'] = df[date_col].dt.dayofweek
	df['is_weekend'] = df[date_col].dt.dayofweek >= 5
	return df


# adicionando as colunas nos dataframes
df_train_full = add_date_features(df_train_full,'date')
df_test_full = add_date_features(df_test_full,'date')
```

```python
df_train_full.head()
```

### Interações entre variáveis

```python
def add_sales_oil_price_interections(df, sales_col, oil_price_col):
  """
  Adiciona interações entre as colunas de vendas e preço do óleo.

  Parâmetros:
  df: DataFrame de entrada dos dados
  sales_col: Nome da coluna de vendas
  oil_price_col: Nome da coluna de preço do óleo

  Returns:
  pd.DataFrame: DataFrame com interações adicionadas
  """
  df['sales_oil_price_interaction_mult'] = df[sales_col] * df[oil_price_col]
  df['sales_oil_price_interaction_div'] = df[sales_col] / df[oil_price_col]
  return df


# adicionando as interações nos dataframes
df_train_full = add_sales_oil_price_interections(df_train_full, 'sales', 'oil_price')
df_test_full = add_sales_oil_price_interections(df_test_full, 'sales', 'oil_price')
```

### Variáveis Lag

```python
# Seleciona as colunas que têm "lag" no nome
lag_cols = [col for col in train_data.columns if 'lag' in col]

# Preenche apenas essas colunas com 0
train_data[lag_cols] = train_data[lag_cols].fillna(0)

# Preenche apenas essas colunas com 0
val_data[lag_cols] = val_data[lag_cols].fillna(0)
```

### Variáveis de médias móveis

```python
media_movel_col = [col for col in train_data.columns if 'rolling' in col]

# Loop pelas colunas de médias móveis
for col in media_movel_col:
    # Preenche primeiro com forward fill por produto
    train_data[col] = (
        train_data
        .groupby('family')[col]
        .ffill()
    )

    val_data[col] = (
        val_data
        .groupby('family')[col]
        .ffill()
    )

    # Depois, preenche o que restar com a média histórica do produto
    train_data[col] = (
        train_data
        .groupby('family')[col]
        .transform(lambda x: x.fillna(x.mean()))
    )

    val_data[col] = (
        val_data
        .groupby('family')[col]
        .transform(lambda x: x.fillna(x.mean()))
    )
```
